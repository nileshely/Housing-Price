```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
housing_data = pd.read_csv("/kaggle/input/housing-price-dataset/Housing.csv")
```

```python
housing_data.head()
```

```python
housing_data.describe()
```

```python
sns.set_style("whitegrid")
```

```python
# Visualization 1: Distribution of Prices
plt.figure(figsize=(10, 6))
sns.histplot(housing_data['price'], bins=30, color='blue', kde=True)
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Prices')
plt.show()
```

```python
# Visualization 2: Number of Bedrooms
plt.figure(figsize=(10, 6))
sns.countplot(x='bedrooms', data=housing_data, palette='Blues_d')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.title('Number of Bedrooms')
plt.show()
```

```python
# Visualization 3: Number of Bathrooms
plt.figure(figsize=(10, 6))
sns.countplot(x='bathrooms', data=housing_data, palette='Blues_d')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Count')
plt.title('Number of Bathrooms')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 4: Scatter plot of Living Area vs Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=housing_data, color='blue')
plt.xlabel('Living Area (sqft)')
plt.ylabel('Price ($)')
plt.title('Living Area vs Price')
plt.show()
```

```python
# Visualization 5: Scatter plot of Lot Size vs Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_lot', y='price', data=housing_data, color='blue')
plt.xlabel('Lot Size (sqft)')
plt.ylabel('Price ($)')
plt.title('Lot Size vs Price')
plt.show()
```

```python
# Visualization 6: Boxplot of Price by Grade
plt.figure(figsize=(10, 6))
sns.boxplot(x='grade', y='price', data=housing_data, palette='Blues_d')
plt.xlabel('Grade')
plt.ylabel('Price ($)')
plt.title('Price by Grade')
plt.show()
```

```python
# Visualization 7: Boxplot of Price by Waterfront
plt.figure(figsize=(10, 6))
sns.boxplot(x='waterfront', y='price', data=housing_data, palette='Blues_d')
plt.xlabel('Waterfront (0: No, 1: Yes)')
plt.ylabel('Price ($)')
plt.title('Price by Waterfront')
plt.show()
```

```python
# Visualization 8: Distribution of Year Built
plt.figure(figsize=(10, 6))
sns.histplot(housing_data['yr_built'], bins=30, color='blue', kde=True)
plt.xlabel('Year Built')
plt.ylabel('Frequency')
plt.title('Distribution of Year Built')
plt.show()
```

```python
# Visualization 9: Distribution of Renovation Year
plt.figure(figsize=(10, 6))
sns.histplot(housing_data['yr_renovated'], bins=30, color='blue', kde=True)
plt.xlabel('Year Renovated')
plt.ylabel('Frequency')
plt.title('Distribution of Renovation Year')
plt.show()
```

```python
# Visualization 10: Count of Properties by Condition
plt.figure(figsize=(10, 6))
sns.countplot(x='condition', data=housing_data, palette='Blues_d')
plt.xlabel('Condition')
plt.ylabel('Count')
plt.title('Count of Properties by Condition')
plt.show()
```

```python
# Visualization 11: Scatter plot of Latitude vs Longitude
plt.figure(figsize=(10, 6))
sns.scatterplot(x='long', y='lat', data=housing_data, color='blue')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Latitude vs Longitude')
plt.show()
```

```python
# Visualization 12: Distribution of Living Area (15 nearest neighbors)
plt.figure(figsize=(10, 6))
sns.histplot(housing_data['sqft_living15'], bins=30, color='blue', kde=True)
plt.xlabel('Living Area of 15 Nearest Neighbors (sqft)')
plt.ylabel('Frequency')
plt.title('Distribution of Living Area (15 nearest neighbors)')
plt.show()
```

```python
# Visualization 13: Distribution of Lot Size (15 nearest neighbors)
plt.figure(figsize=(10, 6))
sns.histplot(housing_data['sqft_lot15'], bins=30, color='blue', kde=True)
plt.xlabel('Lot Size of 15 Nearest Neighbors (sqft)')
plt.ylabel('Frequency')
plt.title('Distribution of Lot Size (15 nearest neighbors)')
plt.show()
```

```python
# Visualization 14: Count of Properties by Zipcode
plt.figure(figsize=(14, 6))
sns.countplot(x='zipcode', data=housing_data, palette='Blues_d')
plt.xlabel('Zipcode')
plt.ylabel('Count')
plt.title('Count of Properties by Zipcode')
plt.xticks(rotation=90)
plt.show()
```
